```python
import scipy as sp
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def van_der_pol(x, t):
    
    mu = 1.0
    
    [x1, x2] = x #x1 -> posicao, x2 = dx1/dt
    
    dxdt = [x2, mu*(1-x1**2)*x2-x1]
    
    return dxdt

def duffing(x, t):
    
    gama = 1
    sigma = .1
    alpha = 1
    beta = 1
    w = 1
    
    [x1, x2] = x
    
    dxdt = [x2, gama*sp.cos(w*t) -sigma*x2 - alpha*x1 - beta*x1**3]
    
    return dxdt
```

```python
#condicoes iniciais
x10 = .1
x20 = 0.1
x0 = [x10, x20] # vetor condicoes iniciais

#tempo
tf = 50
t = sp.linspace(0, tf, 100000)
```

```python
sol1 = odeint(van_der_pol, x0, t)
sol2 = odeint(duffing, x0, t)
```

```python
#Van der Pol
plt.figure()
plt.plot(t,sol1[:,0],'b-', label='x1')
plt.plot(t,sol1[:,1],'r-', label='x2')
plt.legend()
plt.xlabel('t')
plt.ylabel('x')

plt.figure()
plt.plot(sol1[:,0],sol1[:,1])
plt.xlabel('x1')
plt.ylabel('x2')
```

```python
#Duffing
plt.figure()
plt.plot(t,sol2[:,0],'b-', label='x1')
plt.plot(t,sol2[:,1],'r-', label='x2')
plt.legend()
plt.xlabel('t')
plt.ylabel('x')

plt.figure()
plt.plot(sol2[:,0],sol2[:,1])
plt.xlabel('x1')
plt.ylabel('x2')
```
